Task 1: Class and Object (OOP)

Definition
A class is a blueprint used to create objects. An object is an instance of a class that contains actual data and can use the class methods.

Problem Statement
Create a Student class with attributes name and marks. Create one object and display its details.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

s1 = Student("Rahul", 85)
s1.display()
```

Task 2: List

Definition
A list is an ordered and mutable collection of elements.

Problem Statement
Create a list of numbers and calculate the total and average.

```python
numbers = [10, 20, 30, 40]
total = sum(numbers)
average = total / len(numbers)

print("Total:", total)
print("Average:", average)
```

```python
Task 3: Tuple

Definition
A tuple is an ordered collection of elements that cannot be modified after creation.

Problem Statement
Create a tuple of subjects and print all subjects.
```

```python
subjects = ("Math", "Science", "English")
for subject in subjects:
    print(subject)
```

Task 4: Set

Definition
A set is an unordered collection of unique elements.

Problem Statement
Remove duplicate values from a list using a set.

```python
values = [1, 2, 2, 3, 4, 4, 5]
unique_values = set(values)

print("Original List:", values)
print("Unique Values:", unique_values)
```

Task 6: NumPy

Definition
NumPy is a Python library used for numerical computations and array operations.

Problem Statement
Create a NumPy array and calculate mean and standard deviation.

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
```

Task 7: Pandas DataFrame

Definition
A Pandas DataFrame is a two-dimensional table used for storing and analyzing structured data.

Problem Statement
Create a DataFrame containing student names and marks.

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Anita", "Vikram"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)
print(df)
```

Task 8: Filtering in Pandas

Definition
Filtering selects specific rows from a DataFrame based on a condition.

Problem Statement
Display students who scored more than 80 marks.

```python
high_scorers = df[df["Marks"] > 80]
print(high_scorers)
```

Task 10: Matplotlib Visualization

Definition
Matplotlib is a Python library used for creating data visualizations such as graphs and charts.

Problem Statement
Create a bar chart to visualize student marks.

```python
import matplotlib.pyplot as plt

names = ["Rahul", "Anita", "Vikram"]
marks = [85, 90, 78]

plt.bar(names, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
